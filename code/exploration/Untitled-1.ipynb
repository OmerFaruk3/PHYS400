```python
import numpy as np
from scipy.spatial import cKDTree as KDTree
from pesummary.io import read

# ══════════════════════════════════════════════════════════════════════
# 1. VERİ YÜKLEME
# ══════════════════════════════════════════════════════════════════════

file_name = "/Users/omerfaruk/Desktop/PHYS400-Code/PHYS400/Data/"\
            "IGWN-GWTC2p1-v2-GW150914_095045_PEDataRelease_mixed_cosmo.h5"

data = read(file_name, disable_conversion=True)
label = "C01:IMRPhenomXPHM"

posterior_samples = data.samples_dict[label]
prior_samples     = data.priors["samples"][label]

# ══════════════════════════════════════════════════════════════════════
# 2. 2D PARAMETRE SEÇİMİ: mass_1_source ve mass_2_source
# ══════════════════════════════════════════════════════════════════════

params_2d = ["mass_1_source", "mass_2_source"]

# Ham dizileri çek
P_raw = np.column_stack([np.array(posterior_samples[p]) for p in params_2d])
Q_raw = np.column_stack([np.array(prior_samples[p])     for p in params_2d])

print(f"Posterior (P) ham boyut : {P_raw.shape}")
print(f"Prior     (Q) ham boyut : {Q_raw.shape}")

# ── Posterior'dan subsample (k-NN O(n²) yakın olduğu için hız kritik) ──
# 5000 sample istatistiksel olarak yeterli (d=2 için eşik: d²/2 = 2)
N_sub = 5_000
rng   = np.random.default_rng(seed=42)
idx   = rng.choice(len(P_raw), size=N_sub, replace=False)
P_sub = P_raw[idx]          # (5000, 2) — subsampled posterior
Q_sub = Q_raw.copy()        # (5000, 2) — prior zaten 5000 sample

print(f"\nKullanılan boyutlar:")
print(f"  P_sub : {P_sub.shape}")
print(f"  Q_sub : {Q_sub.shape}")

# ══════════════════════════════════════════════════════════════════════
# 3. STANDARDİZASYON — k-NN için ŞAR
# ══════════════════════════════════════════════════════════════════════
# k-NN Öklid uzaklığına dayandığı için parametrelerin birimleri önemlidir.
# mass_1 [~30 M☉] ve mass_2 [~25 M☉] benzer aralıkta ama yine de
# standartlaştırıyoruz; Gaussian KL ise doğrudan orijinal uzayda çalışır.

# Standartlaştırma: prior'un istatistiklerine göre (referans dağılım Q)
mu_q  = np.mean(Q_sub, axis=0)   # (2,)
std_q = np.std(Q_sub,  axis=0)   # (2,)

P_std = (P_sub - mu_q) / std_q   # standardize posterior
Q_std = (Q_sub - mu_q) / std_q   # standardize prior

print(f"\nStandartlaştırma referansı (prior istatistikleri):")
for i, p in enumerate(params_2d):
    print(f"  {p:<20}: μ={mu_q[i]:.2f}, σ={std_q[i]:.2f}")

# ══════════════════════════════════════════════════════════════════════
# 4. YÖNTEM 1 — ANALİTİK GAUSS KL (orijinal uzayda)
# ══════════════════════════════════════════════════════════════════════

def gaussian_kl_2d_bits(P, Q):
    """
    İki çok değişkenli Gaussian arasındaki KL ıraksamasını hesaplar.
    D_KL(P‖Q) = ½ [tr(Σ_Q⁻¹ Σ_P) + (μ_Q-μ_P)ᵀ Σ_Q⁻¹ (μ_Q-μ_P) - d + ln(det Σ_Q / det Σ_P)]
    
    Dikkat: Bu formül P ve Q'nun Gaussian olduğunu varsayar.
    Girdi: Ham (standartlaştırılmamış) sample'lar
    Çıktı: bit cinsinden KL değeri
    """
    d     = P.shape[1]
    mu_p  = np.mean(P, axis=0)
    mu_q  = np.mean(Q, axis=0)
    sig_p = np.cov(P, rowvar=False)
    sig_q = np.cov(Q, rowvar=False)

    inv_sig_q          = np.linalg.inv(sig_q)
    term1              = np.trace(inv_sig_q @ sig_p)
    diff               = mu_q - mu_p
    term2              = diff.T @ inv_sig_q @ diff
    term3              = -d
    _, logdet_p        = np.linalg.slogdet(sig_p)
    _, logdet_q        = np.linalg.slogdet(sig_q)
    term4              = logdet_q - logdet_p

    kl_nats = 0.5 * (term1 + term2 + term3 + term4)
    return kl_nats / np.log(2)

kl_gaussian = gaussian_kl_2d_bits(P_sub, Q_sub)

# ══════════════════════════════════════════════════════════════════════
# 5. YÖNTEM 2 — k-NN KL TAHMİNCİSİ (standartlaştırılmış uzayda)
# ══════════════════════════════════════════════════════════════════════

def knn_kl_bits(x, y):
    """
    Pérez-Cruz (2008) + Wang et al. (2006) k-NN tabanlı KL tahmincisi.
    
    Formül: D_KL(P‖Q) ≈ (d/n) Σ ln(s_i/r_i) + ln(m/(n-1))
    
    r_i : x_i'nin P içindeki en yakın komşu uzaklığı (kendisi hariç, k=2)
    s_i : x_i'nin Q içindeki en yakın komşu uzaklığı (k=1)
    
    Dikkat: Gist'teki kodda işaret düzeltmesi yapılmış.
    Kaynak: https://gist.github.com/atabakd/ed0f7581f8510c8587bc2f41a094b518
    
    Girdi: Standartlaştırılmış sample'lar (birim ölçek önemli!)
    Çıktı: bit cinsinden KL değeri
    """
    x = np.atleast_2d(x)
    y = np.atleast_2d(y)

    n, d  = x.shape
    m, dy = y.shape
    assert d == dy, "Boyut uyuşmazlığı!"

    # KD-Tree oluştur
    xtree = KDTree(x)
    ytree = KDTree(y)

    # r_i: x'in kendi içindeki en yakın komşu (k=2 çünkü k=1 kendisi)
    r = xtree.query(x, k=2, eps=0.01, p=2)[0][:, 1]

    # s_i: x'in y içindeki en yakın komşu (k=1)
    s = ytree.query(x, k=1, eps=0.01, p=2)[0]

    # Wang et al. (2006) Eq. 11 — nats cinsinden
    kl_nats = np.log(s / r).sum() * d / n + np.log(m / (n - 1.))

    return kl_nats / np.log(2)   # bit'e çevir

print("\nk-NN hesaplanıyor... (birkaç saniye sürebilir)")
kl_knn = knn_kl_bits(P_std, Q_std)

# ══════════════════════════════════════════════════════════════════════
# 6. SONUÇLAR VE KARŞILAŞTIRMA
# ══════════════════════════════════════════════════════════════════════

print("\n" + "═"*60)
print(f"  PARAMETRE ÇİFTİ: {params_2d[0]} & {params_2d[1]}")
print("═"*60)
print(f"  Yöntem 1 — Analitik Gaussian KL :  {kl_gaussian:8.4f} bit")
print(f"  Yöntem 2 — k-NN KL tahmincisi   :  {kl_knn:8.4f} bit")
print(f"  Fark (Gauss - kNN)              :  {kl_gaussian - kl_knn:+8.4f} bit")
print(f"  Göreli fark                     :  {abs(kl_gaussian-kl_knn)/kl_knn*100:.2f}%")
print("═"*60)

# Yorumlama rehberi
print("\nYorum:")
print("  • Eğer Gaussian KL > k-NN KL → Gaussian varsayımı KL'yi abartıyor")
print("  • Eğer Gaussian KL < k-NN KL → Gerçek dağılım Gaussian'dan daha")
print("    'karmaşık' (non-Gaussian yapı daha fazla bilgi içeriyor)")
print("  • Fark küçükse (<10%) → mass parametreleri için Gaussian iyi bir")
print("    yaklaşım demektir")
```

```python
import numpy as np
from scipy.stats import gaussian_kde
from pesummary.io import read

# ══════════════════════════════════════════════════════════════════════
# 1. VERİ YÜKLEME
# ══════════════════════════════════════════════════════════════════════

file_name = ("/Users/omerfaruk/Desktop/PHYS400-Code/PHYS400/Data/"
             "IGWN-GWTC2p1-v2-GW150914_095045_PEDataRelease_mixed_cosmo.h5")

data              = read(file_name, disable_conversion=True)
posterior_samples = data.samples_dict["C01:IMRPhenomXPHM"]
prior_samples     = data.priors["samples"]["C01:IMRPhenomXPHM"]

params_2d = ["mass_1_source", "mass_2_source"]

P_raw = np.column_stack([np.array(posterior_samples[p]) for p in params_2d])
Q_raw = np.column_stack([np.array(prior_samples[p])     for p in params_2d])

print(f"Posterior (P): {P_raw.shape}")
print(f"Prior     (Q): {Q_raw.shape}")

# ══════════════════════════════════════════════════════════════════════
# 2. NEDEN DÜZGÜN HISTOGRAM ÇALIŞMIYOR — TANISAL ÇIKTI
# ══════════════════════════════════════════════════════════════════════

print("\n── Aralık Karşılaştırması ──")
for i, p in enumerate(params_2d):
    print(f"  {p}:")
    print(f"    Prior     : [{Q_raw[:,i].min():.2f}, {Q_raw[:,i].max():.2f}]")
    print(f"    Posterior : [{P_raw[:,i].min():.2f}, {P_raw[:,i].max():.2f}]")

# ══════════════════════════════════════════════════════════════════════
# 3. SUBSAMPLE — KDE hesabı için
# ══════════════════════════════════════════════════════════════════════
# KDE bandwidth scott/silverman yöntemi büyük n'de iyi çalışır ama
# 147k sample kde.evaluate() çağrısını yavaşlatır.
# 10_000 sample KDE tahmini için istatistiksel olarak yeterli.

rng   = np.random.default_rng(seed=42)
idx   = rng.choice(len(P_raw), size=10_000, replace=False)
P_sub = P_raw[idx]     # (10000, 2) posterior subsample — KDE için
Q_sub = Q_raw.copy()   # (5000,  2) prior — tamamını kullan

# ══════════════════════════════════════════════════════════════════════
# 4. KDE FİT
# ══════════════════════════════════════════════════════════════════════
# gaussian_kde girdi formatı: (d, n) — satırlar boyut, sütunlar sample
# bw_method='scott' : Scott kuralı, n^(-1/(d+4))
# 2D için bu genellikle iyi çalışır.

print("\nKDE fit ediliyor...")
p_kde = gaussian_kde(P_sub.T, bw_method='scott')
q_kde = gaussian_kde(Q_sub.T, bw_method='scott')

print(f"  Posterior KDE bandwidth faktörü : {p_kde.factor:.4f}")
print(f"  Prior KDE bandwidth faktörü     : {q_kde.factor:.4f}")

# ══════════════════════════════════════════════════════════════════════
# 5. YÖNTEM A — MONTE CARLO KL TAHMİNCİSİ (KDE tabanlı)
# ══════════════════════════════════════════════════════════════════════
# D_KL(P‖Q) = E_P[log p(x) - log q(x)]
#           ≈ (1/n) Σ_i [log p̂(x_i) - log q̂(x_i)]   x_i ~ P
#
# Bu en doğrudan ve varyansı düşük tahmincidir.
# x noktaları zaten P'den geldiği için P'yi yeniden örneklemeye gerek yok.
# logpdf kullanmak numerik taşmayı önler.

print("\nYöntem A: Monte Carlo KL hesaplanıyor...")

log_p_at_P = p_kde.logpdf(P_sub.T)   # log p̂(x_i)
log_q_at_P = q_kde.logpdf(P_sub.T)   # log q̂(x_i)

kl_mc_nats = np.mean(log_p_at_P - log_q_at_P)
kl_mc_bits = kl_mc_nats / np.log(2)

print(f"  KL (Monte Carlo, KDE) = {kl_mc_bits:.4f} bit")

# ══════════════════════════════════════════════════════════════════════
# 6. YÖNTEM B — GRID İNTEGRASYONU (histogram mantığı, KDE ile)
# ══════════════════════════════════════════════════════════════════════
# D_KL(P‖Q) = ∫∫ p(x) log[p(x)/q(x)] dx1 dx2
#           ≈ Σ_{i,j}  p̂(x_{ij}) · log[p̂(x_{ij})/q̂(x_{ij})] · Δx1·Δx2
#
# Grid posterior aralığını kapsar — prior orada ne kadar yoğunsa o kullanılır.
# Bu yaklaşım histogram KL'nin "doğru" versiyonu:
# histogram = kaba KDE, burada sürekli KDE kullanıyoruz.

print("\nYöntem B: Grid entegrasyon KL hesaplanıyor...")

n_grid = 200   # her eksende nokta sayısı → 200×200 = 40,000 nokta

# Grid sınırları: posterior'un min/max'ı + küçük marj
margin = 0.05
m1_min = P_raw[:, 0].min()
m1_max = P_raw[:, 0].max()
m2_min = P_raw[:, 1].min()
m2_max = P_raw[:, 1].max()

m1_range = m1_max - m1_min
m2_range = m2_max - m2_min

m1_grid = np.linspace(m1_min - margin*m1_range, m1_max + margin*m1_range, n_grid)
m2_grid = np.linspace(m2_min - margin*m2_range, m2_max + margin*m2_range, n_grid)

M1, M2   = np.meshgrid(m1_grid, m2_grid)            # (200, 200)
grid_pts = np.vstack([M1.ravel(), M2.ravel()])       # (2, 40000)

# KDE'yi grid üzerinde değerlendir
p_vals = p_kde(grid_pts).reshape(n_grid, n_grid)     # p̂(x)
q_vals = q_kde(grid_pts).reshape(n_grid, n_grid)     # q̂(x)

# Hücre alanı
dx1 = m1_grid[1] - m1_grid[0]
dx2 = m2_grid[1] - m2_grid[0]
dA  = dx1 * dx2

# Normalleştirme kontrolü (her ikisi de yaklaşık 1 olmalı)
p_integral = p_vals.sum() * dA
q_integral = q_vals.sum() * dA

print(f"  p KDE integralı (kontrol, ≈1): {p_integral:.4f}")
print(f"  q KDE integralı (kontrol, ≈1): {q_integral:.4f}")

# KL hesabı — sadece p > 0 olan yerlerde (p=0 → katkı 0)
eps  = 1e-300   # log(0) koruması; q çok küçükse KL büyür ama sonsuz olmaz
mask = p_vals > 0

kl_grid_nats = np.sum(
    p_vals[mask] * np.log(p_vals[mask] / (q_vals[mask] + eps))
) * dA
kl_grid_bits = kl_grid_nats / np.log(2)

print(f"  KL (Grid, KDE) = {kl_grid_bits:.4f} bit")

# ══════════════════════════════════════════════════════════════════════
# 7. REFERANS: GAUSSİAN KL
# ══════════════════════════════════════════════════════════════════════

def gaussian_kl_2d_bits(P, Q):
    d         = P.shape[1]
    mu_p, mu_q   = np.mean(P, axis=0), np.mean(Q, axis=0)
    sig_p, sig_q = np.cov(P, rowvar=False), np.cov(Q, rowvar=False)
    inv_q = np.linalg.inv(sig_q)
    diff  = mu_q - mu_p
    _, ld_p = np.linalg.slogdet(sig_p)
    _, ld_q = np.linalg.slogdet(sig_q)
    kl_nat = 0.5*(np.trace(inv_q @ sig_p) + diff@inv_q@diff - d + ld_q - ld_p)
    return kl_nat / np.log(2)

kl_gauss = gaussian_kl_2d_bits(P_raw, Q_raw)

# ══════════════════════════════════════════════════════════════════════
# 8. SONUÇ TABLOSU
# ══════════════════════════════════════════════════════════════════════

print("\n" + "═"*60)
print(f"  SONUÇLAR: {params_2d[0]} & {params_2d[1]}")
print("═"*60)
print(f"  Yöntem A  Monte Carlo KDE    : {kl_mc_bits:8.4f} bit")
print(f"  Yöntem B  Grid entegrasyon   : {kl_grid_bits:8.4f} bit")
print(f"  Referans  Gaussian KL        : {kl_gauss:8.4f} bit")
print("─"*60)
print(f"  A–B farkı                    : {abs(kl_mc_bits-kl_grid_bits):8.4f} bit")
print(f"  A–Gauss farkı                : {kl_mc_bits - kl_gauss:+8.4f} bit")
print(f"  (+ → Gaussian KL'yi küçümsüyor, dağılım non-Gaussian)")
print(f"  (– → Gaussian KL'yi abartıyor, dağılım Gaussian'a yakın)")
print("═"*60)
```

```python
import numpy as np
from scipy.spatial import cKDTree as KDTree
from pesummary.io import read

# ── 1. VERİ YÜKLEME VE 2D MATRİS OLUŞTURMA ──────────────────────────────
file_name = "/Users/omerfaruk/Desktop/PHYS400-Code/PHYS400/Data/IGWN-GWTC2p1-v2-GW150914_095045_PEDataRelease_mixed_cosmo.h5"

data = read(file_name, disable_conversion=True)
label = "C01:IMRPhenomXPHM"

posterior_samples = data.samples_dict[label]
prior_samples = data.priors["samples"][label]

# Sadece iki kütle parametresi
parameters_2D = ["mass_1_source", "mass_2_source"]

P_cols = []
Q_cols = []

for param in parameters_2D:
    P_cols.append(np.array(posterior_samples[param]))
    Q_cols.append(np.array(prior_samples[param]))

P_2D = np.column_stack(P_cols)  # Posterior: (147634, 2)
Q_2D = np.column_stack(Q_cols)  # Prior:     (5000, 2)

print(f"2D Posterior matrisi (P_2D): {P_2D.shape}")
print(f"2D Prior matrisi (Q_2D):     {Q_2D.shape}")

# ── 2. GİST'TEKİ k-NN KL HESAPLAMASI (Perez-Cruz 2008) ──────────────────
def knn_kl_divergence_bits(x, y):
    """
    Kullback-Leibler divergence estimation using k-Nearest Neighbors.
    Based on: https://gist.github.com/atabakd/ed0f7581f8510c8587bc2f41a094b518
    """
    x = np.atleast_2d(x)
    y = np.atleast_2d(y)

    n, d = x.shape
    m, dy = y.shape

    assert(d == dy)

    # Ağaçları oluştur (Sorgu hızını artırmak için eps=0.01 kullanılıyor)
    xtree = KDTree(x)
    ytree = KDTree(y)

    # x için en yakın 2. komşu (çünkü 1. komşu noktanın kendisidir, mesafe 0 olur)
    r = xtree.query(x, k=2, eps=0.01, p=2)[0][:, 1]
    
    # x noktalarının y içindeki en yakın 1. komşusu
    s = ytree.query(x, k=1, eps=0.01, p=2)[0]

    # Sıfır mesafeleri önlemek için küçük bir tolerans (çakışan noktalar için)
    r = np.maximum(r, 1e-12)
    s = np.maximum(s, 1e-12)

    # Gist'teki formül (Nats cinsinden)
    kl_nats = -np.log(r/s).sum() * (d / n) + np.log(m / (n - 1.))
    
    # Nats -> Bits dönüşümü
    kl_bits = kl_nats / np.log(2)
    
    return kl_bits

# ── 3. ANALİTİK GAUSS KL HESAPLAMASI (Önceki kodun aynısı) ──────────────
def gaussian_kl_divergence_bits(P, Q):
    d = P.shape[1] 
    mu_p = np.mean(P, axis=0) 
    mu_q = np.mean(Q, axis=0) 

    sigma_p = np.cov(P, rowvar=False) 
    sigma_q = np.cov(Q, rowvar=False) 

    inv_sigma_q = np.linalg.inv(sigma_q)  

    term1 = np.trace(np.dot(inv_sigma_q, sigma_p))
    diff = mu_q - mu_p
    term2 = np.dot(diff.T, np.dot(inv_sigma_q, diff)) 
    term3 = -d 
    
    sign_p, logdet_p = np.linalg.slogdet(sigma_p) 
    sign_q, logdet_q = np.linalg.slogdet(sigma_q)
    term4 = logdet_q - logdet_p

    kl_nat = 0.5 * (term1 + term2 + term3 + term4)
    kl_bit = kl_nat / np.log(2)

    return kl_bit

# ── 4. KARŞILAŞTIRMA VE ÇIKTI ───────────────────────────────────────────
print("\n" + "-" * 55)
print("İki Parametre (mass_1, mass_2) İçin KL Karşılaştırması")
print("-" * 55)

print("Hesaplanıyor (k-NN KDTree ile 147k örneklem için birkaç saniye sürebilir)...")

kl_knn = knn_kl_divergence_bits(P_2D, Q_2D)
kl_gauss = gaussian_kl_divergence_bits(P_2D, Q_2D)

print(f"k-NN (Non-parametric / Gist) Yöntemi : {kl_knn:.4f} bit")
print(f"Gauss (Analitik) Yöntemi               : {kl_gauss:.4f} bit")
print("-" * 55)
```

```python
import numpy as np
from scipy.spatial import cKDTree as KDTree
from pesummary.io import read

# ── 1. VERİ YÜKLEME VE 2D MATRİS OLUŞTURMA ──────────────────────────────
file_name = "/Users/omerfaruk/Desktop/PHYS400-Code/PHYS400/Data/IGWN-GWTC2p1-v2-GW150914_095045_PEDataRelease_mixed_cosmo.h5"

data = read(file_name, disable_conversion=True)
label = "C01:IMRPhenomXPHM"

posterior_samples = data.samples_dict[label]
prior_samples = data.priors["samples"][label]

# Sadece iki kütle parametresi
parameters_2D = ["mass_1_source", "mass_2_source"]

P_cols = []
Q_cols = []

for param in parameters_2D:
    P_cols.append(np.array(posterior_samples[param]))
    Q_cols.append(np.array(prior_samples[param]))

P_2D = np.column_stack(P_cols)  # Posterior: (147634, 2)
Q_2D = np.column_stack(Q_cols)  # Prior:     (5000, 2)

# ── 2. GİST'TEKİ k-NN KL HESAPLAMASI (Perez-Cruz 2008) ──────────────────
def knn_kl_divergence_bits(x, y):
    x = np.atleast_2d(x)
    y = np.atleast_2d(y)
    n, d = x.shape
    m, dy = y.shape

    assert(d == dy)

    xtree = KDTree(x)
    ytree = KDTree(y)

    r = xtree.query(x, k=2, eps=0.01, p=2)[0][:, 1]
    s = ytree.query(x, k=1, eps=0.01, p=2)[0]

    r = np.maximum(r, 1e-12)
    s = np.maximum(s, 1e-12)

    kl_nats = -np.log(r/s).sum() * (d / n) + np.log(m / (n - 1.))
    return kl_nats / np.log(2)

# ── 3. ANALİTİK GAUSS KL HESAPLAMASI ────────────────────────────────────
def gaussian_kl_divergence_bits(P, Q):
    d = P.shape[1] 
    mu_p = np.mean(P, axis=0) 
    mu_q = np.mean(Q, axis=0) 

    sigma_p = np.cov(P, rowvar=False) 
    sigma_q = np.cov(Q, rowvar=False) 

    inv_sigma_q = np.linalg.inv(sigma_q)  

    term1 = np.trace(np.dot(inv_sigma_q, sigma_p))
    diff = mu_q - mu_p
    term2 = np.dot(diff.T, np.dot(inv_sigma_q, diff)) 
    term3 = -d 
    
    sign_p, logdet_p = np.linalg.slogdet(sigma_p) 
    sign_q, logdet_q = np.linalg.slogdet(sigma_q)
    term4 = logdet_q - logdet_p

    kl_nat = 0.5 * (term1 + term2 + term3 + term4)
    return kl_nat / np.log(2)

# ── 4. DÜZELTİLMİŞ 2D HİSTOGRAM KL HESAPLAMASI (Laplace Smoothing) ──────
def histogram_kl_divergence_bits_fixed(P, Q, N_bins=100):
    edges = []
    for i in range(P.shape[1]):
        min_val = min(P[:, i].min(), Q[:, i].min())
        max_val = max(P[:, i].max(), Q[:, i].max())
        edges.append(np.linspace(min_val, max_val, N_bins + 1))
        
    # Sadece adetleri (count) alıyoruz
    post_counts, _ = np.histogramdd(P, bins=edges)
    prior_counts, _ = np.histogramdd(Q, bins=edges)
    
    # Laplace Smoothing: Prior'daki sıfır kutularını engellemek için küçük bir değer ekliyoruz
    prior_counts = prior_counts + 1e-5
    
    # Her bir hücrenin alanını/hacmini (dV) hesapla
    dV = np.prod([e[1] - e[0] for e in edges])
    
    # Count'ları Olasılık Yoğunluk Fonksiyonuna (PDF) çeviriyoruz
    post_hist = post_counts / (np.sum(post_counts) * dV)
    prior_hist = prior_counts / (np.sum(prior_counts) * dV)
    
    # Sadece Posterior'un olduğu yerleri hesaplamak yeterli
    mask = post_hist > 0 
    
    kl_nats = np.sum(post_hist[mask] * np.log(post_hist[mask] / prior_hist[mask])) * dV
    return kl_nats / np.log(2)

# ── 5. KARŞILAŞTIRMA VE ÇIKTI ───────────────────────────────────────────
print("\n" + "=" * 65)
print("  İki Parametre (mass_1_source, mass_2_source) Metot Karşılaştırması  ")
print("=" * 65)

print("Hesaplanıyor (k-NN KDTree birkaç saniye sürebilir)...\n")

kl_gauss = gaussian_kl_divergence_bits(P_2D, Q_2D)
kl_knn = knn_kl_divergence_bits(P_2D, Q_2D)
kl_hist = histogram_kl_divergence_bits_fixed(P_2D, Q_2D, N_bins=100)

print(f"1. Gauss (Analitik) Yöntemi               : {kl_gauss:.4f} bit")
print(f"2. k-NN (Non-parametric / Gist) Yöntemi   : {kl_knn:.4f} bit")
print(f"3. Histogram (Laplace Smoothing, 100x100) : {kl_hist:.4f} bit")
print("=" * 65)
```
